```r
# making new column showing ratio of deaths with confirmed cases and confirmed cases with total population in percentage
clean_cases <- educated_cases %>% mutate(C_D_Ratio = (deaths/confirmed_cases) * 100)
```

```r
clean_cases <- clean_cases %>% mutate(Pop_to_cases = (confirmed_cases/total_pop) * 100)
```

```r
# creating a better range for age by adding features and removing the past features for males
# creating ages 0-24
clean_cases$male_0_to_24 <- c(educated_cases$male_under_5 + educated_cases$male_5_to_9 + educated_cases$male_10_to_14 + 
                             educated_cases$male_15_to_17 + educated_cases$male_18_to_19 + educated_cases$male_20 + 
                             educated_cases$male_21 + educated_cases$male_22_to_24)
clean_cases <- subset(clean_cases, select = -c(male_under_5, male_5_to_9, male_10_to_14, male_15_to_17, male_18_to_19, male_20, 
                                               male_21, male_22_to_24))
# creating ages 25-34
clean_cases$male_25_to_34 <- c(educated_cases$male_25_to_29 + educated_cases$male_30_to_34)
clean_cases <- subset(clean_cases, select = -c(male_25_to_29, male_30_to_34))

# creating ages 35-44
clean_cases$male_35_to_44 <- c(educated_cases$male_35_to_39 + educated_cases$male_40_to_44)
clean_cases <- subset(clean_cases, select = -c(male_35_to_39, male_40_to_44))

# creating ages 45-54
clean_cases$male_45_to_54 <- c(educated_cases$male_45_to_49 + educated_cases$male_50_to_54)
clean_cases <- subset(clean_cases, select = -c(male_45_to_49, male_50_to_54))

# creating ages 55-64
clean_cases$male_55_to_64 <- c(educated_cases$male_55_to_59 + educated_cases$male_60_61 + educated_cases$male_62_64)
clean_cases <- subset(clean_cases, select = -c(male_55_to_59, male_60_61, male_62_64))

# creating ages 65-74
clean_cases$male_65_to_74 <- c(educated_cases$male_65_to_66 + educated_cases$male_67_to_69 + educated_cases$male_70_to_74)
clean_cases <- subset(clean_cases, select = -c(male_65_to_66, male_67_to_69, male_70_to_74))

# creating ages 75-84
clean_cases$male_75_to_84 <- c(educated_cases$male_75_to_79 + educated_cases$male_80_to_84)
clean_cases <- subset(clean_cases, select = -c(male_75_to_79, male_80_to_84))

head(clean_cases)
```

```r
# creating a better range for age by adding features and removing the past features for females
# creating ages 0-24
clean_cases$female_0_to_24 <- c(educated_cases$female_under_5 + educated_cases$female_5_to_9 + educated_cases$female_10_to_14 + 
                             educated_cases$female_15_to_17 + educated_cases$female_18_to_19 + educated_cases$female_20 + 
                             educated_cases$female_21 + educated_cases$female_22_to_24)
clean_cases <- subset(clean_cases, select = -c(female_under_5, female_5_to_9, female_10_to_14, female_15_to_17, female_18_to_19,
                                               female_20, female_21, female_22_to_24))
# creating ages 25-34
clean_cases$female_25_to_34 <- c(educated_cases$female_25_to_29 + educated_cases$female_30_to_34)
clean_cases <- subset(clean_cases, select = -c(female_25_to_29, female_30_to_34))

# creating ages 35-44
clean_cases$female_35_to_44 <- c(educated_cases$female_35_to_39 + educated_cases$female_40_to_44)
clean_cases <- subset(clean_cases, select = -c(female_35_to_39, female_40_to_44))

# creating ages 45-54
clean_cases$female_45_to_54 <- c(educated_cases$female_45_to_49 + educated_cases$female_50_to_54)
clean_cases <- subset(clean_cases, select = -c(female_45_to_49, female_50_to_54))

# creating ages 55-64
clean_cases$female_55_to_64 <- c(educated_cases$female_55_to_59 + educated_cases$female_60_to_61 + educated_cases$female_62_to_64)
clean_cases <- subset(clean_cases, select = -c(female_55_to_59, female_60_to_61, female_62_to_64))

# creating ages 65-74
clean_cases$female_65_to_74 <- c(educated_cases$female_65_to_66 + educated_cases$female_67_to_69 + educated_cases$female_70_to_74)
clean_cases <- subset(clean_cases, select = -c(female_65_to_66, female_67_to_69, female_70_to_74))

# creating ages 75-84
clean_cases$female_75_to_84 <- c(educated_cases$female_75_to_79 + educated_cases$female_80_to_84)
clean_cases <- subset(clean_cases, select = -c(female_75_to_79, female_80_to_84))

head(clean_cases)
```
